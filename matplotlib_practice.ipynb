```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the Plot
plt.plot(x_axis, y_axis)
```

```python
# Create a Figure using the OO Method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Alternative, the above cell could be coded as follows:
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0, 45)
ax.grid()
ax.legend()
```

## 5.1.7 Create Scatter Plots and Bubble Charts

```python
# *** Create a Scatter Plot Using the MATLAB Method *** #

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.legend()
```

```python
# *** Create a Bubble Chart Using the MATLAB Method *** #

# Create a list for scatter to iterate through the axes in order to set 
# the sizes to a multiple of 3
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*10)
    
# Now pass the array to the scatter plot
# Turn the scatter plot above into a bubble chart
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use Python's "List Comprehension" technique to "loop through" the list
# more efficently than using a for loop
# ex: new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# *** This is used to send color names vs. rbg values to set the color w/ or w/o transparency
import matplotlib.colors as colors

fig, ax = plt.subplots()

# c to set color
# edgecolors to set the color of the edge to black
# linewidth to set width of edge
# alpha to set percentage of transparency # *** NOTE: this sets all colors transparent
# 
# ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c="skyblue", ec="k", 
#             lw=2, alpha=0.2, label="Boston")

# *** This works to keep the border from being transparent
# NOTE: fc=(1,0,0,0.5), ec=(0,0,0,1) will set the colors using r, b, g, alpha
# to use "skyblue" instead of the value between 0-1 for b(lue)
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], fc=colors.to_rgba("skyblue", 0.2), ec=colors.to_rgba("k"),
            lw=2, label="Boston")

ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)
ax.invert_yaxis()
```
